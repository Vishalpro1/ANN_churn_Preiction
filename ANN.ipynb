```python
data = pd.read_csv("/kaggle/input/annchurn-modelling/Churn_Modelling.csv")
```

```python
data.sample(5)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Geography'] = le.fit_transform(data['Geography'])
data['Gender'] = le.fit_transform(data['Gender'])
```

```python
data
```

```python
data.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
data
```

```python
x = data.iloc[:, :10] # Features are columns from index 0 up to (but not including) 10
y = data.iloc[:, 10]  # Target variable 'Exited' is at index 10
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# BUILDING THE MODEL
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# initializing the ANN model
model =  Sequential()
```

```python
# adding the input layer and the first hidden layer
model.add(Dense(16, activation = 'relu', input_shape = (10,)))
```

```python
# adding the second hidden layer
model.add(Dense(16, activation = 'relu'))

# adding the output layer for binary classification
model.add(Dense(1, activation = 'sigmoid'))
```

```python
# compiling the ANN
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
# fitting the ANN
model.fit(x_train, y_train, batch_size = 10, epochs = 100)
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
